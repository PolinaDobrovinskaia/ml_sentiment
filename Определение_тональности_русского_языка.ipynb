# Пример использования библиотеки Hugging Face для анализа тональности русского текста 

Устанавливает библиотеку Hugging Face (модуль transformers)

```python
!pip install transformers
```

Создаем пайплайн для определения тональности текста

```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis", "blanchefort/rubert-base-cased-sentiment")
```

Используем пайплайн для определения тональности текста

```python
classifier("Я ненавижу API и программную инженерию")
```

```python
classifier("Я обожаю программировать на Python")
```

```python
classifier("Я обожаю читать книги")
```
